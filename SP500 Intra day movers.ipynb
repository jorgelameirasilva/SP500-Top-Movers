```python
import yfinance as yf
import pandas as pd
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
```

```python
sp500_tickers = pd.read_html(url)
```

```python
tickers = sp500_tickers[0]
```

```python
tickers
```

```python
main_df = pd.DataFrame(columns=["Ticker", "Movement"])

for ticker in tickers["Symbol"]:
    df = yf.download(ticker, period="5d")
    movement = (df["High"] - df["Low"]) / df["Low"]
    movement = movement.mean()
    main_df = main_df.append({"Ticker":ticker, "Movement":movement}, ignore_index=True)
```

```python
main_df.dropna(inplace=True)
```

```python
biggest_movers = main_df.sort_values(by="Movement",ascending=False).head(10)
```

```python
biggest_movers
```

```python
additional_data = tickers[["Symbol", "Security", "GICS Sector"]]
```

```python
additional_data
```

```python
final_df = biggest_movers.merge(additional_data, left_on="Ticker", right_on="Symbol")
```

```python
final_df.drop("Symbol", axis=1, inplace=True)
```
